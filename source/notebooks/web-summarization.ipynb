```python
import pandas as pd
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vQZ2S0QjGtV4XIEOdUQvtFC1aI45OPTtOA0bwhFrpjVn1DmHOrfG1OCCRtKgKqJ0Af18660LAC96xII/pubhtml/sheet?headers=false&gid=0#gid=0'
data = pd.read_html(url, header=1, encoding='utf8')[0]
data = data[data['Website'].notna()]
data['Founded'] = data['Founded'].astype(str).str.replace('.0', '', regex=False)
data
```

```python
import numpy as np
from helpsk.pandas import relocate

def clean_raised(value: str) -> float:
    if value.startswith('$'):
        return float(value[1:].replace('+', ''))
    elif 'undisclosed' in value.lower():
        return np.nan
    else:
        return np.nan

data['raised_millions'] = data['Total Raised (M)'].apply(lambda x: clean_raised(x))
data = relocate(data, column='raised_millions', before='Total Raised (M)')
data[['Company', 'raised_millions', 'Total Raised (M)', 'Founded']].\
    sort_values('raised_millions', ascending=False).\
    reset_index(drop=True).\
    style.\
    bar()
```

```python
website = data.sort_values('raised_millions', ascending=False).iloc[1]['Website']
website
```

```python
# from langchain.document_loaders import UnstructuredURLLoader
from langchain.document_loaders import SeleniumURLLoader

website = data.sort_values('raised_millions', ascending=False).iloc[3]['Website']
print(website)
if 'www.' in website:
    website = website.replace('www.', 'http://')
if 'http' not in website:
    website = 'http://' + website

website += '/careers'

print(website)
loader = SeleniumURLLoader(urls=[website])
content = loader.load()
content
```

```python
def add_newlines(string):
    for i in range(0, len(string), 100):
        print(string[i:i+100])

add_newlines(content[0].page_content)
```
